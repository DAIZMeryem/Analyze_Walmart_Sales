#**DATA CLEANING**

###*Diplicated records*

```python
data.duplicated().sum()
```

we have 51 duplicated records to delete

```python
data.drop_duplicates(inplace=True)
data.duplicated().sum()
```

###*Missing values*

```python
data.isnull().sum()
```

we have 31 records that the haven't unit_price or quantity so I will delete them because they will not help in analysing the data.

```python
data.dropna(inplace=True)
data.isnull().sum()
```

```python
data.shape
```

###*Fix the Inappropriate type*

```python
data["unit_price"]=data["unit_price"].str.replace("$","").astype(float)
```

```python
data.describe()
```

###*Add significate mesures*

```python
data["total_price"]=data["unit_price"]*data["quantity"]
```

#*Change the columns name to lower case*

```python
data.columns=data.columns.str.lower()
```

```python
data.to_csv("Walmart_cleaned.csv",index=False)
```
